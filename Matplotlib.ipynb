```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
var1 = np.arange(0, 10)
```

```python
var1
```

```python
var2 = var1**2
```

```python
var2
```

```python
plt.plot(var1, color = 'red', label = 'Var1')
plt.plot(var2, color = 'green', label = 'Var2')
plt.xlabel('Time')
plt.ylabel('Prices')
plt.title('Time vs Price')
plt.legend(title = 'Graphs')
```
