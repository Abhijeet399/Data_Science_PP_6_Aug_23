```python
path = 'C:/Users/bhatt/OneDrive/Desktop/whisper-main/sherlock/'

def read_all_data(path):
    text = []
    
    for _, _, files in os.walk(path):
        for file in files:
            with open(path + file) as f:
                for line in f:
                    line = line.strip()
                    if line == '----------':
                        break
                    if line !='':
                        text.append(line)
    return text
```

```python
all_data = read_all_data(path)
```

```python
len(all_data)
```

```python
def clean_data(lines):
    clean_text = []
    for line in lines:
        line = line.lower()
        line = re.sub(r"[-()\"#/@;:<>{}`+=~|.!?,]", "", line)
        tokens = word_tokenize(line)
        words = []
        for word in tokens:
            words.append(word)
        clean_text += words
    return clean_text
```

```python
cleaned_lines = clean_data(all_data)
```

```python
len(cleaned_lines)
```

```python
cleaned_lines
```

```python
n_gram = 2

markov_model = {}

for i in range(len(cleaned_lines) - n_gram - 1):
    curr_state, next_state = "", ""
    for j in range(n_gram):
        curr_state += cleaned_lines[i + j] + " "
        next_state += cleaned_lines[i + j + n_gram] + " "
    curr_state = curr_state[:-1]
    next_state = next_state[:-1]
    
    if curr_state not in markov_model:
        markov_model[curr_state] = {}
        markov_model[curr_state][next_state] = 1
    else:
        if next_state in markov_model[curr_state]:
            markov_model[curr_state][next_state] += 1
        else:
            markov_model[curr_state][next_state] = 1
print(markov_model)
```

```python
for curr_state, next_state in markov_model.items():
#     total = 0
#     for i in next_state.values():
#         total += i
    total = sum(next_state.values())
#     print(total)
    for state, count in next_state.items():
        markov_model[curr_state][state] = count/total
```

```python
import random

def generate_lines(markov_model, start, limit = 10):
    
    count = 0
    curr_state = start
    next_state = None
    line = ""
    line += curr_state + " "
    
    while count < limit:
        next_state = random.choices(list(markov_model[curr_state].keys()),
                                   list(markov_model[curr_state].values()))
        count += 1
#         print(next_state)
        curr_state = next_state[0]
        line += curr_state + ' '
    return line
```

```python
generate_lines(markov_model, 'i am')
```
