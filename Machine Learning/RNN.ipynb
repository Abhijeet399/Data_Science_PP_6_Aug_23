```python
import tensorflow as tf
import numpy as np
import keras
from tensorflow.keras.datasets import imdb
```

```python
(train_X, train_Y), (test_X, test_Y) = imdb.load_data()
```

```python
print(train_X.shape, train_Y.shape, test_X.shape, test_Y.shape)
```

```python
len(train_X[20])
```

```python
train_Y[1]
```

```python
max_len = 400
train_X = tf.keras.preprocessing.sequence.pad_sequences(train_X, maxlen = max_len)
test_X = tf.keras.preprocessing.sequence.pad_sequences(test_X, maxlen = max_len)
```

```python
print(train_X.shape, train_Y.shape, test_X.shape, test_Y.shape)
```

```python
max_num_input = np.max(train_X) + 1
vec_len = 50
model = keras.Sequential()
```

```python
model.add(keras.layers.Embedding(max_num_input, vec_len))
model.add(keras.layers.SimpleRNN(100, dropout = 0.2))
model.add(keras.layers.Dense(1, activation = 'sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer = 'Adam', loss = 'binary_crossentropy', metrics = 'accuracy')
```

```python
model.fit(train_X, train_Y, epochs = 500, batch_size = 100)
```
