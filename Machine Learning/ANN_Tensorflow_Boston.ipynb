```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from keras.datasets import boston_housing
```

```python
(train_X, train_Y), (test_X, test_Y) = boston_housing.load_data()
```

```python
print(train_X.shape, train_Y.shape, test_X.shape, test_Y.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
min_max_scaler = MinMaxScaler()
```

```python
min_max_scaler.fit(train_X)
train_X = min_max_scaler.fit_transform(train_X)
```

```python
min_max_scaler.fit(test_X)
test_X = min_max_scaler.fit_transform(test_X)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential(name = 'ANN_Boston_Price_Prediction')
```

```python
model.add(Dense(188, input_dim = train_X.shape[1], activation = 'relu', name = 'Pratham_Layer'))
model.add(Dense(130, activation = 'relu', name = 'Doosra_Layer'))
model.add(Dense(95, activation = 'relu', name = 'Teesra_Layer'))
model.add(Dense(45, activation = 'relu', name = 'Chotha_Layer'))
model.add(Dense(1, activation = 'linear', name = 'Lakshay_Layer'))
```

```python
model.summary()
```

```python
model.compile(optimizer = 'Adam', loss = 'mse', metrics = ['mae'])
```

```python
hist = model.fit(train_X, train_Y, batch_size = 50, epochs = 300, validation_split = 0.1)
```

```python
plt.plot(hist.history['loss'][200:])
plt.plot(hist.history['val_loss'][200:])
```

```python
plt.plot(hist.history['mae'])
plt.plot(hist.history['val_mae'])
```

```python
model.evaluate(test_X, test_Y)
```

```python
preds = model.predict(test_X)
```

```python
for i in range(len(preds)):
    print(preds[i], test_Y[i])
```
