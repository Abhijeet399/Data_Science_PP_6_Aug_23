```python
def hypothesis(x, weight):
    y_pred = 0
    for i in range(x.shape[0]):
        y_pred += weight[i] * x[i]
    return y_pred
```

```python
def loss_func(x, y, weight):
    
    error = 0
    for i in range(x.shape[0]):
        
        y_pred = hypothesis(x[i], weight)
        
        error += (y_pred - y[i])**2
    return error/x.shape[0]
```

```python
def gradient(x, y, weight):
    
    grad = np.zeros((x.shape[1], ))
    
    for j in range(x.shape[1]):
        for i in range(x.shape[0]):

            y_pred = hypothesis(x[i], weight)

            grad[j] += (y_pred - y[i]) * x[i][j]
        
    return grad/x.shape[0]
```

```python
def gradient_descent(x, y, learning_rate = 0.03):
    
    weight = np.zeros((x.shape[1], ))
    
    cost = []
    
    epoch = 6000
    
    for i in range(epoch):
        grad = gradient(x, y, weight)
        
        loss = loss_func(x, y, weight)
        cost.append(loss)
        for j in range(x.shape[1]):
            weight[j] = weight[j] - learning_rate * grad[j]
        
    return cost, weight
```

```python
def gradient_descent(x, y, learning_rate = 0.03):
    
    weight = np.zeros((x.shape[1], ))
    
    cost = []
    
    epoch = 6000
    
    for i in range(epoch):
        grad = gradient(x, y, weight)
        
        loss = loss_func(x, y, weight)
        cost.append(loss)

        weight = weight - learning_rate * grad
        
    return cost, weight
```
