```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from mpl_toolkits.mplot3d import Axes3D
```

```python
data, label = make_circles(n_samples = 2000, noise = 0.02, random_state = 5)
```

```python
plt.scatter(data[:, 0], data[:, 1], c = label)
```

```python
print('Diameter =', max(data[:, 0]) - min(data[:, 0]))
```

```python
data.shape
```

```python
axis_1 = data[:, 0].reshape(-1, 1)
axis_2 = data[:, 1].reshape(-1, 1)
axis_3 = (axis_1**2 + axis_2**2)

X = np.hstack((data, axis_3))
```

```python
X.shape
```

```python
axes = plt.subplot(projection = '3d')
axes.scatter(axis_1, axis_2, axis_3, c = label)
plt.show()
```

```python
from sklearn import svm

svc = svm.SVC(kernel = 'linear')
svc.fit(X, label)
```

```python
svc.score(X, label)
```

```python
from sklearn import svm

svc = svm.SVC(kernel = 'poly', degree = 250)
svc.fit(X, label)
print(svc.score(X, label))
```
